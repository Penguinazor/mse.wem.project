```python
# https://arxiv.org/help/api/index
# https://arxiv.org/help/api/user-manual

import xml.etree.ElementTree as ET
from urllib.request import urlopen

# http://export.arxiv.org/api/query?search_query=all:electron&start=0&max_results=1

base_url = 'http://export.arxiv.org/api/query?search_query='
categroy = "stat.ML"
subject = "electron"
start = 1
max_results = 0

url = base_url + 'all:' + subject + '&start=' + str(start) + '&max_results=' + str(max_results)
data = urlopen(url).read()
parsed = ET.fromstring(data)

#print(parsed)

#for child in parsed:
#    print(child.tag)
#    print(child.attrib)

#print(parsed.findall('{http://www.w3.org/2005/Atom}entry'))


for e in parsed.findall('{http://www.w3.org/2005/Atom}entry'):
    for i in e.findall('{http://www.w3.org/2005/Atom}author'):
        print(i[0].text)


#print(parsed[0:])

#for e in parsed:
#    print(e)

for e in parsed[7:]:
    for s in e:
        if s.text == None:
            print(s)
        else:
            print(s.text)
            
#print(e[5][0].text)
#print(e[5][0].text)

#print(parsed.getroot())
#print(parsed.getElementsByTagName('authors'))
#print(itemlist[0].attributes['name'].value)
```
